```python
import spacy
from collections import Counter
import csv
import pygal
from pygal.style import DarkStyle
```

```python
spacy.cli.download("en_core_web_lg")
nlp = spacy.load('en_core_web_lg')

            
```

```python
file_list1 = ["og-memo-scrape.txt", "game-dialogue.txt"] 
file_list2 = ["mul-scrape.txt", "eraserhead-script.txt", "jacobs-ladder-script.txt", "shinning-dialogue.txt"]
```

```python
def posCollector(SHpos, words):
    wordList = []
    count = 0
    print(f"Finding {SHpos} in text entry...")
    for token in words:
        if token.pos_ == f"{SHpos}":
            count += 1
            wordList.append(token.lemma_)
    return(wordList)
```

```python
for file in file_list1:
    re_file = open(f"../../text-files/{file}", "r", encoding="utf-8").read()
    spacyRead = nlp(re_file)
    print(f"Reading {file}")
    
    SHpos = posCollector("ADJ", spacyRead)
    wordFreq = Counter(SHpos)
    topCommon = wordFreq.most_common()[:15]

    numba = []
    for word, freq in wordFreq.most_common():
        numba.append(freq)
        
    print(sum(numba)) 

    with open(f"output/ajd_{file}.txt", "w") as f:
        for word, freq in wordFreq.most_common()[:40]:
            f.write(f"{word}: {freq}\n")
    

for file in file_list2:
    re_file = open(f"../../text-files/{file}", "r", encoding="utf-8").read()
    spacyRead = nlp(re_file)
    print(f"Reading {file}")
    
    SHpos = posCollector("ADJ", spacyRead)
    wordFreq = Counter(SHpos)
    topCommon = wordFreq.most_common()[:15]

    numba = []
    for word, freq in wordFreq.most_common():
        numba.append(freq)
        
    print(sum(numba)) 

    with open(f"output/adj_{file}", "w") as f:
        for word, freq in wordFreq.most_common()[:40]:
            f.write(f"{word}: {freq}\n")


bar_chart = pygal.HorizontalBar()
bar_chart.title = 'Adjectives to Total word count'
bar_chart.add('Game Dialogue', 513)
bar_chart.add('Game Memos', 350)
bar_chart.add('The Shining', 1129)
bar_chart.add('Jacobs Ladder', 1850)
bar_chart.add('Mulholland Dr', 331)
bar_chart.add('Eraserhead', 52)
bar_chart.render()


bar_chart.render_to_file(f'output/adj_{file}.svg')
```
