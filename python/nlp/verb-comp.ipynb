### POS Counter

```python
# Creating Part of Speech Collector
def posCollector(whapos, words):
    wordList = []
    count = 0
    print(f"Finding {whapos} in text entry...")
    for token in words:
        if token.pos_ == f"{whapos}":
            count += 1
            wordList.append(token.lemma_)
    return(wordList)
```

```python
for file in file_list:
    re_file = open(f"../../text-files/{file}", "r", encoding="utf-8").read()
    spacyRead = nlp(re_file)
    print(f"Reading {file}")
    
    whapos = posCollector("VERB", spacyRead)
    wordFreq = Counter(whapos)
    topCommon = wordFreq.most_common()[:15]

    with open(f"output/vrbs_{file}.txt", "w") as f:
        for word, freq in wordFreq.most_common()[:40]:
            f.write(f"{word}: {freq}\n")
    
    pie_chart = pygal.Pie(style=DarkStyle, inner_radius=0.4)
    pie_chart.title = f"Verb Composition of {file} (in qty)"

    # Put total data in chart
    for word, freq in topCommon:
        pie_chart.add(word, freq)

    pie_chart.render_to_file(f'output/vrbs_{file}.svg')

# Lexical diversity for Verbs (Check later. fairly high? but with high amt of thematic verbs)
# One analysis for most common and another for lexical diversity
```
