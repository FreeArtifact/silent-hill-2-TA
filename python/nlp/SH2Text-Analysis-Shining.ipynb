Silent Hill and The Shining Comparative Text Analysis

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import FreqDist
from nltk.text import Text
import matplotlib.pyplot as plt

with open("../../text-files/shinning-dialogue.txt", "r", encoding="utf-8") as f:
    shining_text = f.read()

with open("../../text-files/game-dialogue.txt", "r", encoding="utf-8") as f:
    sh_text = f.read()

shining_tokens = word_tokenize(shining_text.lower())
sh_tokens = word_tokenize(sh_text.lower())
```

Frequency Distribution

```python
shining_freq = FreqDist(shining_tokens)
sh_freq = FreqDist(sh_tokens)

print("\nTop 40 words in The Shining:\n")
print(shining_freq.most_common(20))

print("\nTop 40 words in Silent Hill 2:\n")
print(sh_freq.most_common(20))
```

Shared Vocabulary

```python
shining_vocab = set(shining_tokens)
sh_vocab = set(sh_tokens)

common_words = shining_vocab.intersection(sh_vocab)

print("\nSome shared words:\n")
print(list(common_words)[:50])
```

Themes

```python
themes = ["crazy", "kill", "love", "wife", "alive"]

print("\nTheme counts:\n")
for word in themes:
    print(f"{word}:")
    print("  Shining:", shining_freq[word])
    print("  SH2:", sh_freq[word])
```

Dispersion plots

```python
shining_text_nltk = Text(shining_tokens)
sh_text_nltk = Text(sh_tokens)

words_to_track = ["suffering", "kill", "alive", "nightmare", "love", "wife"]

print("\nShowing dispersion plot for The Shining...")
shining_text_nltk.dispersion_plot(words_to_track)
plt.show()

print("\nShowing dispersion plot for Silent Hill 2...")
sh_text_nltk.dispersion_plot(words_to_track)
plt.show()
```
