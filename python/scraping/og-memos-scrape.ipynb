```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
base_url = "https://www.silenthillmemories.net/sh2/memos_en.htm"

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )
}
```

```python

# stored line by line
# with open("../../text-files/og-memo-scrape.txt", "w") as f:
    
# Response request
response = requests.get(base_url, headers=headers)

# Telemetry
if response.status_code != 200:
    print(f"Failure! Status Code: {response.status_code}")
else:
    print(f"Success! Status Code: {response.status_code}")

# Parse content for page
tree = BeautifulSoup(response.content, "html.parser")

divs = tree.find_all("div", class_="div_desc")

with open("../../text-files/og-memo-scrape.xml", "w") as f:
    for div in divs:
        f.write("<div>\n")
        # div children because some text has no p element
        for child in div.children:
            # 
            if child.name is None:
                empty_text = child.strip()
                # In case there are empty elements
                if empty_text:
                    f.write(f"<p>{empty_text}</p>\n")
            # if the p has a class then surround it in maintext element
            elif child.name == "p" and len(child.attrs) >= 1:
                f.write(f"<mainText>{child.get_text(' ', strip=True)}</mainText>\n")
            elif child.name == "p":
                # Surround each p with element and join those stripped elements with " "
                f.write(f"<p>{child.get_text(' ', strip=True)}</p>\n")
        f.write("</div>\n")
```
