```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
base_url = "https://www.scripts.com/script/mulholland_drive_14210"

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )
}
```

```python
def scrape_pages(start_page, end_page):
    # stored line by line
    full_script = []
    for page_num in range(start_page, end_page + 1):
        # Current page url
        if page_num == 0:
            url = base_url
        else:
            url = f"{base_url}{page_num}"
        print(f"Attempting Scrape on Page: {page_num}")
        
        # Response request
        response = requests.get(url, headers=headers)

        # Telemetry
        if response.status_code != 200:
            print(f"Failure! Status Code: {response.status_code}")
            break
        else:
            print(f"Success! Status Code: {response.status_code}")

        # Parse content for page
        tree = BeautifulSoup(response.content, "html.parser")

        container = tree.find("div", id="disp-quote-body")
        for script in container:
            script_page = container.find("p")
            print(script.get_text(strip=True))
            line = script.get_text(strip=True)
            full_script.append({
                'page':page_num,
                'line':line
            })

        time.sleep(0.5)
    return full_script

```
